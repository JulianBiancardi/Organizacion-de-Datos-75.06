* Generar los siguientes dataframes conteniendo:

```
import pandas as pd
import numpy as np
```

```
df_ventas = pd.read_csv("/DataSets/ventas.csv")
df_ventas.dropna(inplace=True)
df_ventas
```

###Promedio por producto


```
productos_agrupados = df_ventas.groupby('codigo_producto').agg({'ventas_mensuales':'mean'})
productos_agrupados.rename(columns = {'ventas_mensuales' : 'promedio_ventas'},inplace=True)
productos_agrupados.head()
```

### Mínimo, Máximo y Promedio de ventas por producto 

Primero generamos el formato general para los 3 casos

```
productos_agrupados = df_ventas.groupby('codigo_producto').agg({'ventas_mensuales': ['min','max','mean']})

column_rename = {'min':'Minimo ventas mensual','max':'Maximo ventas mensual','mean':'Promedio ventas mensual'}
productos_agrupados.rename(columns = column_rename,inplace = True)
productos_agrupados.columns = productos_agrupados.columns.get_level_values(1)
productos_agrupados.head()
```

Para el caso A nos queda:

```
productos_agrupados.stack()
```

Para el caso B nos queda exactamente lo mismo.

Para el caso C nos queda:

```
#Usamos el operador T para intercambiar filas por columnas
productos_agrupados.T
```
